```python
import fitz  # PyMuPDF (for PDFs)
from transformers import pipeline
from google.colab import files  # For file upload/download
import re

# Load Hugging Face AI model for text generation
# model = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.1")
model = pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B")
```

```python
def get_similar_words(search_texts):
    """Use LLM to find synonyms & related words for each inputted word."""
    all_terms = []  # Store original words + AI-generated words

    for text in search_texts:
        prompt = f"Give me synonyms and related words for: {text}"
        result = model(prompt, max_length=50, do_sample=True, truncation=True)  # Uses Hugging Face LLM

        words = result[0]['generated_text'].split(", ")
        all_terms.extend([text] + words)  # Add original and AI-generated words

    return list(set(all_terms))  # Remove duplicates

def highlight_text_in_pdf(pdf_path, search_texts, output_path):
    """Highlights user-inputted words and AI-generated synonyms in the PDF, ensuring whole-word matching."""
    doc = fitz.open(pdf_path)  # Open the PDF
    extracted_text = ""  # Store extracted text for console display

    # Use LLM to get AI-generated similar words
    search_terms = get_similar_words(search_texts)
    print(f"üîπ Searching for: {search_terms}")

    for page_num, page in enumerate(doc):  # Loop through pages
        page_text = page.get_text("text")  # Extract text
        extracted_text += f"\n\n--- Page {page_num + 1} ---\n{page_text}"

        for term in search_terms:
            # Use regex to find whole words only
            pattern = rf'\b{re.escape(term)}\b'  # \b ensures exact word match
            matches = re.finditer(pattern, page_text)

            for match in matches:
                rects = page.search_for(match.group())  # Find word positions
                for rect in rects:
                    page.add_highlight_annot(rect)  # Highlight found text

    doc.save(output_path)  # Save modified PDF
    print(f"\n‚úÖ Highlighted PDF saved as: {output_path}\n")

    # Print extracted text with highlights in yellow (console)
    for term in search_terms:
        extracted_text = re.sub(rf'\b{re.escape(term)}\b', f"\033[43m{term}\033[0m", extracted_text)

    print("üîπ Extracted Text with Highlights:")
    print(extracted_text)

# üìå Upload PDF in Colab
# uploaded = files.upload()  # Upload file manually

# Get the file name of the uploaded PDF
pdf_file = list(uploaded.keys())[0]
text_to_highlight = input("Enter words to highlight (comma-separated): ").split(",")  # Accept multiple words
text_to_highlight = [word.strip() for word in text_to_highlight]  # Remove spaces

output_pdf = "highlighted_output.pdf"

# Run the function
highlight_text_in_pdf(pdf_file, text_to_highlight, output_pdf)

# üì• Download the highlighted PDF
# files.download(output_pdf)
```
